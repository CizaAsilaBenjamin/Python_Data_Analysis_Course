```python

import ast
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datasets import load_dataset
```

```python
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notnull(x) else x)
```

```python
df_DA_US = df[(df['job_title_short']=='Data Analyst') & (df['job_country']=='United States')].copy()
```

```python
df_plot= df_DA_US["job_location"].value_counts().head(10).to_frame()
sns.set_theme(style="ticks")
sns.barplot(data=df_plot, x='count', y='job_location', hue='count', palette='dark:b_r', legend=False)
sns.despine()
plt.title('Top 10 locations for Data Analyst jobs in the US')
plt.xlabel('Number of jobs')
plt.ylabel('')
plt.show()
```

```python
dict_columns = {
    'job_work_from_home': 'Work From Home Offered',
    'job_no_degree_mentioned': 'Degree Requirement',
    'job_health_insurance': 'Health Insurance Offered',
}

fig, ax = plt.subplots(1, 3),
figsize=(15, 5)

for i, (col, title) in enumerate(dict_columns.items()):
    df_plot = df_DA_US[col].value_counts().to_frame()
    sns.barplot(data=df_plot, x=df_plot.index.astype(str), y=col, palette='dark:b_r', ax=ax[i])
    ax[i].set_title(title)
    ax[i].set_xlabel('')
    ax[i].set_ylabel('Number of jobs')
    for p in ax[i].patches:
        ax[i].annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()),
                       ha='center', va='center', fontsize=11, color='black', xytext=(0, 5),
                       textcoords='offset points')
```
