```python
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset['train'].to_pandas()
df.job_posted_date = pd.to_datetime(df['job_posted_date'])
df.pivot_table(values=['salary_year_avg', 'salary_hour_avg'], index='job_country', columns='job_title_short', aggfunc='median')

```

```python
top_countries = df['job_country'].value_counts().head(6).index
```

```python
df_job_country_salary = df.pivot_table(
values='salary_year_avg',
index='job_country',
columns='job_title_short',
aggfunc='median'
)
df_job_country_salary = df_job_country_salary.loc[top_countries]
df_job_country_salary
```

```python
job_title= ['Data Analyst', 'Business Analyst', 'Data Engineer']
df_job_country_salary = df_job_country_salary[job_title]
df_job_country_salary.plot(kind='bar')
plt.ylabel('Salary average ($)')
plt.xlabel()
plt.title('Average salary by countries and jobs')
plt.xticks(rotation=45, ha='rigth')
plt.show()
```
